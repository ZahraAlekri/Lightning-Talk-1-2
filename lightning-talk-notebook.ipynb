### Data Files

The dataset came with 5 files. city_day has daily air quality averaged per city (29,531 rows). station_day has daily readings per individual monitoring station (108,035 rows). city_hour and station_hour are the hourly versions of the same data. stations is just metadata about each station like name and coordinates.

The two daily files were merged vertically since they share the same structure and columns. The only difference was the location column name, City in one and StationId in the other, both were renamed to Location before merging. This gave a combined dataset of 137,566 rows.

---

### Target Column

AQI Bucket is the target, it categorizes air quality into 6 levels: Good, Satisfactory, Moderate, Poor, Very Poor and Severe. This is a classification problem. Moderate (34%) and Satisfactory (28%) are the most common classes. Good and Severe are both around 6%. The classes are somewhat imbalanced but no single class is extreme enough to be an immediate problem.

---

### Feature Types

13 continuous numerical columns covering pollutant readings: PM2.5, PM10, NO, NO2, NOx, NH3, CO, SO2, O3, Benzene, Toluene, Xylene and AQI. One categorical column which is Location and one datetime column which is Date.

Note: AQI is the numerical score that AQI Bucket is derived from. Including it as a feature would cause data leakage so it would need to be dropped before modeling.

---

### Missing Values

Xylene is missing 75% of its values. NH3 is missing 42% and PM10 is missing 39%. Toluene is 34% missing and Benzene is 27%. O3 and SO2 are both around 21% missing. PM2.5 is 19% missing despite being one of the strongest predictors of AQI. The target column AQI Bucket is itself 19% missing.

---

### Distributions

Every pollutant column is right skewed. The bulk of daily readings are low with a long tail of very high values on certain days. This is typical for pollution data since most days are normal but occasionally extreme pollution events pull the distribution out. The median is more representative of a typical day than the mean in this case since the mean gets pulled towards those extreme readings.

---

### Correlations

PM10 has the strongest correlation with AQI at 0.87 and PM2.5 follows at 0.78. CO and NO2 are both around 0.51. O3 is the weakest at 0.18. Among the features themselves, NO and NOx are highly correlated at 0.86 since NOx is essentially a combination of NO and NO2. PM2.5 and PM10 are also highly correlated at 0.86 since both measure particulate matter. Benzene and Toluene have a moderate correlation of 0.53.

---

### Outliers

PM2.5 and PM10 both hit a maximum of 1000 which looks like a sensor cap rather than a real reading. AQI reaches 2049, O3 reaches 963 and Benzene reaches 455. These extreme values are visible in the histograms as long tails far from the bulk of the data.

# Topic 3 : Solar Power Generation

## Problem Statement

Inconsistent solar power output creates challenges for grid management, as energy suppliers need to balance supply and demand in real time. Without accurate predictions of how much power a plant will generate, grid operators risk either overloading the network or failing to meet demand. Using inverter-level generation data and plant sensor readings, **can we predict AC power output to support better energy planning and grid stability?**

## Data Info
Data Source : https://www.kaggle.com/datasets/anikannal/solar-power-generation-data

## Data Dictionary
| Column | Type | Description |
|--------|------|-------------|
| DATE_TIME | datetime | Timestamp of the reading, recorded every 15 minutes |
| PLANT_ID | int | Unique identifier for the solar power plant |
| SOURCE_KEY | str | Unique identifier for the inverter within the plant |
| DC_POWER | float | DC power generated by the inverter in kilowatts |
| AC_POWER | float | AC power output after conversion from DC in kilowatts |
| DAILY_YIELD | float | Cumulative energy generated by the inverter for that day in kWh |
| TOTAL_YIELD | float | Total cumulative energy generated by the inverter since installation in kWh |
| AMBIENT_TEMPERATURE | float | Temperature of the surrounding air at the plant in Celsius |
| MODULE_TEMPERATURE | float | Temperature of the solar panel module itself in Celsius |
| IRRADIATION | float | Amount of solar irradiation hitting the panels in kW/m² |

```python
gen1 = pd.read_csv('data/dataset_3/Plant_1_Generation_Data.csv')
gen2 = pd.read_csv('data/dataset_3/Plant_2_Generation_Data.csv')
sen1 = pd.read_csv('data/dataset_3/Plant_1_Weather_Sensor_Data.csv')
sen2 = pd.read_csv('data/dataset_3/Plant_2_Weather_Sensor_Data.csv')
```

```python
for name, d in [('gen1', gen1), ('gen2', gen2), ('sen1', sen1), ('sen2', sen2)]:
    print(f"\n{name}: {d.shape}")
    print(d.columns.tolist())
```

```python
gen1['DATE_TIME'] = pd.to_datetime(gen1['DATE_TIME'], format='%d-%m-%Y %H:%M')
sen1['DATE_TIME'] = pd.to_datetime(sen1['DATE_TIME'])
gen2['DATE_TIME'] = pd.to_datetime(gen2['DATE_TIME'])
sen2['DATE_TIME'] = pd.to_datetime(sen2['DATE_TIME'])
```

```python
plant1 = pd.merge(gen1, sen1[['DATE_TIME', 'AMBIENT_TEMPERATURE', 'MODULE_TEMPERATURE', 'IRRADIATION']], 
                  on='DATE_TIME', how='left')
```

```python
plant2 = pd.merge(gen2, sen2[['DATE_TIME', 'AMBIENT_TEMPERATURE', 'MODULE_TEMPERATURE', 'IRRADIATION']], 
                  on='DATE_TIME', how='left')
```

```python
df3 = pd.concat([plant1, plant2], ignore_index=True)
```
